**DAILY CHECK LIST**

```python
daily_checklist = [ "study","complete Assignement" ," Go to market" , "Playing" , "Running", "Call Friends"]
```

```python
completed_task =[]
incompleted_task =[]
print("End Of day")
```

```python
for task in daily_checklist:
  answer = input(f"Did You complet '{task}' ? (Yes/No):").strip().lower()
  if answer == "yes":
    completed_task.append(task)
  else:
    incompleted_task.append(task)
```

```python
print("\n Completed Task:")
if completed_task:
  for task in completed_task:
    print(f"-{task}")
else:
  print("No task completed Today")
```

```python
print("\n Incompleted Task:")
if incompleted_task:
  for task in incompleted_task:
    print(f"-{task}")
else:
  print("No task completed Today")
```
