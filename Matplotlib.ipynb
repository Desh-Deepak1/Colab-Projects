```python
import matplotlib.pyplot as plt
```

```python
# HOURS VS MARKS
# HOURS : 2,24, 4, 6, 8,9,10
#Mark : 21 , 38 ,64 , 85
```

```python
hours = [2,4,6,8,10]
math_marks =[21, 38, 64 ,85 ,92]
sci_marks= [10,15,36,54,45]
```

```python
plt.scatter(hours , math_marks , label ="math")
plt.scatter(hours , sci_marks , label="Science")
plt.xlabel('hours')
plt.ylabel('marks')
plt.title('hours vs marks')
plt.legend()
plt.show()
```
