### For Loop

A `for` loop is used for iterating over a sequence (that is, a list, a tuple, a dictionary, a set, or a string). It's also commonly used with the `range()` function to iterate a specific number of times.

```python
# Example 1: Iterating over a range of numbers
print("For loop using range:")
for i in range(5): # This will iterate from 0 to 4
    print(i)

# Example 2: Iterating over a list
print("\nFor loop iterating over a list:")
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)
```

### While Loop

A `while` loop executes a set of statements as long as a condition is true. It's important to make sure the condition eventually becomes false to avoid an infinite loop.

```python
# Example: Counting up to 5
print("While loop example:")
count = 0
while count < 5:
    print(count)
    count += 1 # Increment count to eventually make the condition false
```

### Do-While Loop (Simulated in Python)

Python does not have a built-in `do-while` loop. A `do-while` loop executes the loop body *at least once* before checking the condition. You can simulate this behavior in Python using a `while True` loop with an `if` condition and a `break` statement inside.

```python
# Example: Simulating a do-while loop
print("Do-while loop simulation:")
i = 0
while True:
    print(f"Current value of i: {i}")
    i += 1
    if i >= 3: # This is the condition check at the 'end' of the loop
        break # Exit the loop if condition is met

# Another way to think about simulating do-while:
i = 0
print("\nDo-while loop simulation (alternative):")
# Execute at least once
print(f"Current value of i: {i}")
i += 1
# Then continue with a while loop
while i < 3:
    print(f"Current value of i: {i}")
    i += 1
```
