**Q1 . Lets say  we have a List L1 = [1,2,2,3,2,3,4,5] , now identify the most repetative elament in the list**

```python
# 1 - 1
# 2 - 3
# 3 - 2
# 4 - 1
# 5 - 1
```

```python
L1=[1,2,3,2,4,3,5,2]
```

```python
num_counter ={}

# 1:1 , 2:3 , 3:2 , ....
for i in L1:
  if i in num_counter:
    num_counter[i] += 1
  else:
    num_counter[i]=1
```

```python
if 1 in num_counter:
  print('1 is present')
else:
  num_counter[1]=1
```

```python
num_counter
```

```python
temp =0
final_result = None
for k , v in num_counter.items():
  print(f"K : {k} , V :{v}, Temp : {temp}")
  if v>temp:
    temp = v
    final_result =k
```
